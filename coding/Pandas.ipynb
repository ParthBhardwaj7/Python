```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)#np.linspace(0, 10, 100): Generates 100 evenly spaced values between 0 and 10.
y=np.sin(x)
z=np.cos(x)
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z,label="cos(x)")
plt.legend()# it display a small boxe in graph that label the plotted line based on thier label
plt.show()
```

```python
#OR
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sinvalue')
plt.title('sinwave')
plt.show()
```

```python
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
plt.plot(x,y,'g.')
plt.show()
plt.plot(x,y,'rx')
plt.show()
```

```python

# Code	Color	   Marker
# 'r+'	Red	Plus (+)
# 'g.'	Green	Dot (.)
# 'rx'	Red	X     (x)
```
